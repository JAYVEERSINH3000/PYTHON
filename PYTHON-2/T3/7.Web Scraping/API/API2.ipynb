# post requests

```python
import requests
import json
new_post={'title':'Learning requests',
         'body':'This is Tutorial',
         'userId':1}
url="https://jsonplaceholder.typicode.com/posts"
response = requests.post(url,json=new_post)
print(response.json())
print(f"Status code : {response.status_code}")
```

# put requests

```python
import requests
import json
updated_post={'id':1,   # if we don't write id then it will automatilcally asigned id
              'title':'Updated Title',
             'body':'Updated Body',
             'userId':1}
url="https://jsonplaceholder.typicode.com/posts/1"
response = requests.put(url,json=updated_post)
print(response.json())
print(f"Status code : {response.status_code}")
```

# patch request

```python
import requests
import json
updated_post={'title':'Patched Title'}  # only update given data
url="https://jsonplaceholder.typicode.com/posts/1"
response = requests.patch(url,json=updated_post)
print(response.json())
print(f"Status code : {response.status_code}")
```

# delete request

```python
import requests
import json
url="https://jsonplaceholder.typicode.com/posts/1"
response = requests.get(url)
data = response.json()
with open('posts.json','w') as f:
    json.dump(data,f,indent=5)
print("Json downloaded")
```

```python
with open('posts.json','r') as f:
    load_data=json.load(f)
print(load_data)
```

# Nested json content

```python
import requests
import json
import pandas as pd
url="https://jsonplaceholder.typicode.com/users/1"
response = requests.get(url)
user = response.json()
print(json.dumps(user,indent=5))
```

```python
print(f"Username : {user['username']}")
print(f"Email : {user['email']}")
print(f"City : {user['address']['city']}")
print(f"Lattitude : {user['address']['geo']['lat']}")
print(f"Longitude : {user['address']['geo']['lng']}")
print(f"Phone : {user['phone']}")
print(f"Company name : {user['company']['name']}")
```

```python
json_string = '{"name":"Jayveer","Age":20}'
data = json.loads(json_string)
print(data)
print(type(json_string))
print(type(data))
```

```python
import requests
url = "https://jsonplaceholder.typicode.com/posts"
params = {
    '_page':2,
    '_limit':5
}
response = requests.get(url,params=params)
data=response.json()
print(json.dumps(data,indent=5))
```

```python
for post in data:
    print(post['id'],"---",post['title'])
```

```python
import requests
url = "https://jsonplaceholder.typicode.com/posts"
page = 1
while page<=3:
    params = {'_page':page,'_limit':3}
    response = requests.get(url,params=params)
    data=response.json()
    for post in data:
        print(post['id'],"---",post['title'])
    page+=1
```
