```python
l=[]
for i in range(1,11):
    n=int(input("Enter 10 no.="))
    l.append(n)
print(f"max no of list={max(l)}")
print(f"min no of list={min(l)}")
l.sort()
print(f"second max={l[-2]}")
print(f"second min={l[1]}")
even=[]
odd=[]
for j in l:
    if j%2==0:
        even.append(j)
    else:
        odd.append(j)
print(f"even max number={max(even)}")
print(f"even min number={min(even)}")
print(f"odd max number={max(odd)}")
print(f"odd min number={min(odd)}")
```

```python
#longest common subsequent sub-string
x=["pyt","pytho","python","pyth"]
x.sort(reverse=True)
ans=" "
for i in range(len(x[-1])):
    if x[0][i]==x[-1][i]:
            ans=ans+x[i][i]
            continue
    else:
        break
if ans=="":
    print("string not found")
else:
    print(f"longest common subsequence={ans}")
```

```python
l=[2,1,6,4]
e=[]
o=[]
for i,j in enumerate(l):
    if i%2==0:
        e.append(j)
    else:
        o.append(j)
print("even index",e)
print("odd index",o)
```

```python
l=[2,7,6,4]

for i in range(len(l)):
    x=l.copy()
    z=x.pop(i)
    sum1=sum(x[::2])
    sum2=sum(x[1::2])
    if sum1==sum2:
        print("Removable:",z) 
```
