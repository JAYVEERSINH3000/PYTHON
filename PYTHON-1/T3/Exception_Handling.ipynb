```python
try:
    a=10
    b=0
    print(a/b)
except:
    print(a)
finally:
    print("finally")
```

```python
try:
    f=open("word.txt")
except Exception as e:
    print(e)
else:
    x=f.read()
    print(x)

# else execute only when try block is successfully executed.
```

```python
try:
    a=10
    b=0
    print(a/b)
except ZeroDivisionError:
    print("Z")
except NameError:
    print("N")
```

```python
class Demo(Exception):
    pass
try:
    n=input("Enter mobile number :")
    if len(n)==10 and n[0]=="9":
        print("Success")
    elif len(n)!=10 :
        raise Demo("Please enter 10 digit mobile num ")
    else :
        raise Demo("First num must be 9")
except Demo as d:
    print(d)
```
