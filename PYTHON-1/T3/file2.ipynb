```python
#Q1:read the file mbox-short.txt and find all the keyword X-DSPAM-Confidence: 0.8475 
#   from all the file and find average of the value:
#   output: answer is : 0.7507185
f=open("mbox-short.txt")
x=f.readlines()
c=0
s=0
for i in x:
    if "X-DSPAM-Confidence:" in i:
        c=c+1
        p=i.split(" ")
        s=s+float(p[1])
f.close()
print(f"Answer is {s/c}")
```

```python
#Q2:read the file mbox-short.txt and find all the line which start with "From " 
#   o/p : no. of email is : 27
c1=0
for i in x:
    if "From " in i:
        c1=c1+1
        k=i.split(" ")
        print(k[1])
print(f"no. of emails is : {c1}")
```

```python
# Q3 extract emails and how many times its used and print maximum times used email and count : 
#   o/p : no. of email is : 27
c1=0
l=[]
dic={}
for i in x:
    if "From " in i:
        c1=c1+1
        k=i.split(" ")
        l.append(k[1])
        if k[1] in l:
            dic.update({k[1]:l.count(k[1])})
m=max(dic.values())
for i,j in dic.items():
    print(i,j)
for i1,j1 in dic.items():
    if j1==m:
            print(f"Email : {i1} , {m} times")
```
