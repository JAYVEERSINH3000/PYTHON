```python
import pandas as pd
```

```python
a=[1,7,2]
myvar=pd.Series(a)
print(myvar)
print()
print(type(myvar))
print()
print(myvar[0])
```

```python
a=[1,7,2]
myvar=pd.Series(a,index=['x','y','z'])
print(myvar)
print()
print(type(myvar))
print()
print(myvar[0])
print(myvar['y'])
```

```python
a=(1.5,2,3,4,5)
myvar=pd.Series(a)
print(myvar)
```

```python
student_name=['ram','shyam','radhe','seeta','geeta']
marks=[50,60,70,80,90]
myvar=pd.Series(marks,index=student_name,name='student result')
print(myvar)
```
