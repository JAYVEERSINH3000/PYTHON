```python
#Single level Inheritance :
class A:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        print(self.a,self.b)
class B(A):
    def __init__(self,z):
        self.z=z
        print(self.z)
b=B(30)
```

```python
class A:
    def __init__(self):
        print("Parent constructor")
    def show(self):
        print("Parent show")
class B(A):
    def __init__(self):
        print("Child constructor")
        super().__init__()
    def add(self):
        print("Add")
b=B()
b.add()
b.show()
```

```python
# Multi    level inheritance
class A:
    def __init__(self):
        print("cons A")
    def show(self):
        print("hello")
class B(A):
    def __init__(self):
        print("cons B")
    def show(self):
        print("hiii")
class C(B):
    def __init__(self):
        print("cons C")
    def show(self):
        print("Namaste")
ob=C()
ob.show()
```

```python
# Multiple Inheritance
class A:
    def __init__(self):
        print("con A")
class B:
    def __init__(self):
        print("con B")
class C(A,B):
    def __init__(self):
        super().__init__()
        print("Con c")
ob=C()
```

```python
class A:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        print(self.a,self.b)
class B(A):
    def __init__(self,c,a,b):
        super().__init__(a,b)
        self.c=c
        print(self.c)
ob=B(10,20,30)
```

```python
class Student:
    def __init__(self,mark):
        self.mark=mark
    def __add__(self,other):
        return self.mark+other.mark
ob1=Student(70)
ob2=Student(100)
ob1+ob2
```
