# 2-D  DataFrame

```python
import pandas as pd
data={'calories':[420,380,390,400],'duration':[50,40,45,48]}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.loc[0])  # loc used for all type of index
print("")
print(df.loc[1])
```

```python
print(df.iloc[0:2:,0])  #iloc[row,column]  iloc used only for integer index
```

```python
df.loc[[0,2]]
```

```python
# Example 1 :
import pandas as pd
data={'Name':['Alice','Bob','Charlie'] , 'Age':[25,30,35] , 'City' : ['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.loc[0])
print()
df.iloc[0]
```

```python
print(df.loc[0:2])  # End point include
print()
print(df.iloc[0:2]) # End point not include/exclude
```

```python
df.loc[(df.Age>26)]
```

```python
df.loc[(df.Name=='Bob')]
```

```python
df.loc[(df.Age>26) & (df.City=='Chicago')]
```

```python
# Example 2: 
import pandas as pd
data={'A':[1,2,3,4],'B':[5,6,7,8]}
df=pd.DataFrame(data,index=['r1','r2','r3','r4'])
print(df)
print()
df.loc['r2':'r4','B']=99
print(df)
print()
df.iloc[1:3,0]=0
print(df)
df['C']=[11,22,33,44]
print(df)
```

```python
df.loc[df['A']==0,'new_col']='Zero'
df
```

```python
df.loc[len(df)]=[10,11,12,13]
df
```

```python
# Example 3 :
import pandas as pd
data={'Name':['Alice','Bob','Charlie'] , 'Age':[25,30,35] , 'City' : ['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data)
print(df)
```

```python
df.set_index('Name',drop=False,inplace=True)  # inplace : for update dataframe 
df
```
