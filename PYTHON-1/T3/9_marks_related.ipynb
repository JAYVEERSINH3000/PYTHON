```python
# Read file1 and when "#" is appear than suppose after # is comment ,Remove all comments and write in file2
f1=open("file1.txt")
r1=f1.readlines()
f2=open("file2.txt","w+")
r2=f2.read()
print(r1)
for i in r1:
    if i.startswith("#"): # or if i[0]=="#"
        continue
    else:
        if "#" in i:
            ind=i.find("#")
            f2.write(i[0:ind]+"\n")
        else:
            f2.write(i)
f1.close()
f2.close()
```

```python
# Que1 type1
f=open("Auction.txt")
r=f.readlines()
l1=[]
l2=[]
for i in r:
    z=i.split(" ")
    l1.append(z[0])
    l2.append(int(z[1]))
print(l1)
print(l2)
```

```python
d=list(zip(l1,l2))
di={}
for i,j in d:
    di[i]=j
```

```python
m=max(di.values())
mi=min(di.values())
```

```python
for i,j in di.items():
    if j==m:
        print(i,j)
```
