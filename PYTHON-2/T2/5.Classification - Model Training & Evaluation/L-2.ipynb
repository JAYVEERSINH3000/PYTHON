* AptitudeCommunication.csv :-

```python
import pandas as pd
df=pd.read_csv('AptitudeCommunication.csv')
df
```

```python
x=df.iloc[:,1:3]
y=df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=7)
model=nn.fit(x,y)
prediction=model.predict(x)
print(prediction)
```

```python
diff=pd.DataFrame({'actual':y,'predicted':prediction})
diff
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y,prediction)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(y,prediction))
```

```python
print("Error Rate:",1-accuracy_score(y,prediction))
```

# DecisionTree_Sports.csv :

```python
import pandas as pd
df=pd.read_csv('DecisionTree_Sports.csv')
df
```

```python
df.drop('Day',axis=1,inplace=True)
df
```

```python
import numpy as np
df['Weather']=np.where(df['Weather']=='Sunny',2,np.where(df['Weather']=='Rain',1,0))
df
```

```python
df['Temperature']=np.where(df['Temperature']=='Hot',1,np.where(df['Temperature']=="Cool",0,2))
df
```

```python
df['Humidity']=np.where(df['Humidity']=='High',0,1)
df
```

```python
df['Wind']=np.where(df['Wind']=='Strong',0,1)
df
```

```python
df['Play_Sports?']=np.where(df['Play_Sports?']=='Yes',1,0)
df
```

```python
x=df.iloc[:,:4]
y=df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
x.columns
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='entropy')
model=dtree.fit(x,y)
prediction=model.predict(x)
print(prediction)
```

```python
from sklearn import tree
text_representation = tree.export_text(dtree)
print(text_representation)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("Confusion Matrix: ",confusion_matrix(y,prediction))
print("Accuracy Score: ",accuracy_score(y,prediction))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(90,30))
plot_tree(dtree, feature_names=x.columns,filled=True,rounded=True)
plt.show()
```
