Adding a New Feature

```python
import pandas as pd
#defining fetures
room_length=[18,20,10,12,18,11]
room_breadth=[20,20,10,11,19,10]
room_type=['Big','Big','Normal','Normal','Big','Normal']

#creating a dataframe
data=pd.DataFrame({"Length":room_length,"Breadth":room_breadth,"Type":room_type})
data
```

```python
#Adding a feature called area from Length and Breadth
data["Area"]=data['Length']*data["Breadth"]
data
```

```python
data[data["Area"]>200]
```

Encoding Nominal Variables :

```python
import pandas as pd
apartment_area=[4720,2430,4368,3969,6142,7912]
apartment_price=[2360000,1215000,2184000,1984000,3071000,3956000]
data2=pd.DataFrame({"Area":apartment_area,"Price":apartment_price})
data2
```

```python
import numpy as np
data2["Price"]=np.where(data2["Price"]>3000000,"High",
                       np.where(data2["Price"]<2000000,"Low","Medium"))
data2
```
