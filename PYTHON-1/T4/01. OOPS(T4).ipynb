```python
from abc import ABC,abstractmethod
class Employee(ABC):
    def getdata(self):
        i=int(input("Enter employee id : "))
        name=input("Enter employee name : ")
        sal=int(input("Enter employee salary : "))
        self.i=i
        self.name=name
        self.sal=sal
    def display(self):
        print(f"Employee ID: {self.i}")
        print(f"Employee Name: {self.name}")
        print(f"Employee Basic Salary: {self.sal}")
    @abstractmethod
    def emp_id(self):
        pass
class Perks(Employee):
    def cal1(self):
        self.da=(self.sal)*(0.35)
        self.hra=(self.sal)*(0.17)
        self.pf=(self.sal)*(0.12)
        self.total=self.da+self.hra-self.pf
    def emp_id(self):
        pass
    def show(self):
        print(f"DA: {self.da}")
        print(f"HRA: {self.hra}")
        print(f"PF: {self.pf}")
       
class Netsalary(Perks):
    def cal2(self):
        self.fsal=self.sal+self.total
    def show1(self):
        super().getdata()
        super().cal1()
        super().show()
        self.cal2()
        print(f"Total salary: {self.fsal}")
n=Netsalary()
n.show1()
```

```python
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def calculate_area(self):
        return self.l*self.b
class Circle(Shape):
    def __init__(self,r):
        self.radius=r
    def calculate_area(self):
        return (3.14)*(self.radius**2)
r=Rectangle(10,20)
c=Circle(3)
ra=r.calculate_area()
ca=c.calculate_area()
print(f"Area of ractangle is: {ra}\nArea of circle is: {ca}")
```
