```python
import pandas as pd
df=pd.DataFrame({"Hours studied":[1,2,3,4,5,6,7],"Exam score":[50,55,65,70,75,80,85]})
df
```

```python
import numpy as np
# or also make like this
# x=np.array([[1],[2],[3],[4],[5],[6],[7]])
# y=np.array([50,55,65,70,75,80,85])
# print(x)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,0:1]  # or x=df[ ["Hours studied"] ]
y=df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y_pred)
```

```python
diff=pd.DataFrame({"Actual":y,"Predicted":y_pred})
diff
```

```python
print("Coeddicient: ",lr.coef_)
print("Intercept: ",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y,y_pred))
print("MSE: ",metrics.mean_squared_error(y,y_pred))
print("R2 Score:",metrics.r2_score(y,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='blue',label='Actual Scores')
plt.plot(x,y_pred,color='red',label='Predicted Scores')
plt.xlabel('Hours studied')
plt.ylabel('Exam score')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("car data.csv")
df
```

Data Preprocessing

```python
df['Age']=2026-df['Year']
df
```

```python
df=df.drop(['Year'],axis=1)
df
```

```python
df=df.drop('Car_Name',axis=1)
df
```

```python
df=pd.get_dummies(data=df,drop_first=True)
df
```

```python
df.columns
```

Spliting data into train ans test subsets

```python
y=df['Selling_Price']
x=df.drop('Selling_Price',axis=1)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
diff=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
diff
```

```python
print("Cofficient: ",lr.coef_)
print("Intercept: ",lr.intercept_)  
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score:",metrics.r2_score(y_test,y_pred))
```
