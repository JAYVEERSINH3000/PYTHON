# Random module

```python
from numpy import random
```

```python
n=int(input("Enter choice 1-odd / 2-even : "))
x=random.randint(100)
print(f"The number is : {x}")
if n==1:
    if x%2==1:
        print("Won")
    else:
        print("Lose")
else:
    if x%2==0:
        print("Won")
    else:
        print("Lose")
```

```python
import time
k=5
while k>0:
    x=random.randint(100)
    print(x)
    time.sleep(3)
    k=k-1
```
