# Heatmap

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
sns.heatmap(data)
plt.show()
```

```python
hm=sns.heatmap(data,cmap="tab20")  #game te name lakho error ave etle niche options avi jase 
plt.show()
```

```python
hm=sns.heatmap(data,annot=True)  
plt.show()
```

```python
hm=sns.heatmap(data,vmin=10,vmax=70,cmap="tab20",annot=True,linewidths=3,linecolor='black',cbar=True,
               xticklabels=True,yticklabels=False)
plt.show()
```

```python
import pandas as pd
df=pd.DataFrame(np.random.random((5,5)),
               columns=['a','b','c','d','e'])
print(df)
p1=sns.heatmap(df)
```

```python
plt.figure(figsize=(15,5))
data=np.random.randint(low='1',high='100',size=(3,10))
print(data)
hm=sns.heatmap(data=data,vmin=10,vmax=70,cmap="Pastel1",annot=True,linewidths=2,linecolor='black')
```
