```python
import requests
from datetime import datetime
response = requests.get('https://api.github.com/rate_limit')
data = response.json()
print("Limit : ",data['rate']['limit'])
print("Used : ",data['rate']['used'])
print("Left : ",data['rate']['remaining'])
print("Reset : ",data['rate']['reset'])

reset_stamp = data['rate']['reset']
reset_time = datetime.fromtimestamp(reset_stamp).strftime("%y-%m-%d %H:%M:%S")
print(reset_time)
```
