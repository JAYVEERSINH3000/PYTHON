```python
import requests,json
def five_day_forecast():
    api_key='811e06a1f973ce4847ab761c6fca3d22'
    city=input("Enter City Name : ")
    request_url = f" http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    response = requests.get(request_url)
    coordinates = response.json()
    #print(json.dumps(coordinates,indent=5))
    lattitude = coordinates[0]['lat']
    longitude = coordinates[0]['lon']
    
    request_url1= f" http://api.openweathermap.org/data/2.5/forecast?lat={lattitude}&lon={longitude}&appid={api_key}"
    response1 = requests.get(request_url1)
    forecast = response1.json()
    print(json.dumps(forecast,indent=5))
    return forecast,city
fc,city=five_day_forecast()
```

```python
date=[]
time=[]
temp=[]
pressure=[]
humidity=[]
desc=[]
visi=[]
wind=[]
for i in range(0,40):
    dt=fc['list'][i]['dt_txt']
    date.append(dt[0:10])
    time.append(dt[11:])
    t=fc['list'][i]['main']['temp']
    temp.append(t)
    p=fc['list'][i]['main']['pressure']
    pressure.append(p)
    h=fc['list'][i]['main']['humidity']
    humidity.append(h)
    d=fc['list'][i]['weather'][0]['description']
    desc.append(d)
    v=fc['list'][i]['visibility']
    visi.append(v)
    w=fc['list'][i]['wind']['speed']
    wind.append(w)
```

```python
import pandas as pd
df=pd.DataFrame({"Date":date,
               "Time":time,
                "Temperature":temp,
                 "Pressure":pressure,
                "Humidity":humidity,
                "Description":desc,
                "Visibility":visi,
                "Wind Speed":wind})
df
```
