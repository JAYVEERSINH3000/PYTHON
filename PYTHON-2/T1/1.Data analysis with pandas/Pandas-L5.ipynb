# apply()

```python
import pandas as pd
s=pd.Series([35,67,90,45])
def pass_fail(x):
    return 'pass' if x>=50 else 'fail'
result=s.apply(pass_fail)
result
```

```python
df=pd.DataFrame({"physics":[60,70,80],"chemistry":[65,75,85],"maths":[70,80,90]})
result1=df.apply(sum,axis=1) 
print(result1)
```

```python
result2=df.apply(sum,axis=0)
print(result2)
```

```python
result3=df.apply(lambda x:x.mean())
print(result3)
```

```python
df['Total']=df.apply(lambda x:x.sum(),axis=1)
df
```

```python
def grade(Total):
    if Total>=250:
        return "A"
    elif Total>=200:
        return "B"
    else:
        return "C"
```

```python
df["Grade"]=df["Total"].apply(grade)
df
```

```python
import pandas as pd
df=pd.DataFrame({"Width":[1,2],"Height":[3,4]})
def get_stats(row):
    return [row['Width']*row['Height'],2*(row['Width']+row['Height'])]
df
```

```python
df.apply(get_stats,axis=1,result_type='expand')
```

```python
df.apply(get_stats,axis=1,result_type='reduce')
```

```python
df.apply(get_stats,axis=1,result_type='broadcast')
```

```python
import pandas as pd
def add_value(x,y,z):
    return x+y+z
df=pd.DataFrame({'A':[1,2],'B':[10,20]})
```

```python
df1=df.apply(add_value,args=(1,2))
print(df1)
```

```python
def add_value(x,m):
    return x*m
df1=df.apply(add_value,m=10)
print(df1)
```
