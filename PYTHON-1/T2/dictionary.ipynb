```python
x="hi hello how are you i'm fine are you fine"
y=x.split(" ")
d={}
for i in x:
    if d.get(i)==None:
        val=[]
        count=0
        for j in y:
            if j[0]==i:
                val.append(j)
                count+=1
        if count>0:
            d.update({i:val})
    else:
        continue
print(d)        
```
